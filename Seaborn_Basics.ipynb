```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# load dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Creating scatter plot using Figure level function
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', hue='sex', size = 'total_bill', style='smoker')
```

```python
# Creating scatter plot using Axis level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size = 'total_bill', style='smoker')
```
